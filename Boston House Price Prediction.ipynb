```python
from scipy import stats
from scipy.stats import norm,skew
sns.distplot(boston_df['medv'],fit=norm,color="Green")
(mu,sigma) = norm.fit(boston_df['medv'])
plt.legend(["Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})".format(mu,sigma)],loc='best')
plt.ylabel("Frequency")
plt.xlabel("SalesPrice")
plt.title("Sales Price Distribution")
sns.distplot(boston_df['medv'],fit=norm)
fig=plt.figure()
res=stats.probplot(boston_df['medv'],plot=plt)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x=boston_df.drop("medv",axis=1)
y=boston_df["medv"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
predictions=lr_model.predict(x_test)
print("Actual value of the house:-", y_test.iloc[0])
print("Predicted  value of the house:-", predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions)
rmse=np.sqrt(mse)
print("Mean squared error is :-",mse)
print( "root Mean squared error is :-",rmse)
```

```python
lr_model.score(x_test,y_test)
```
